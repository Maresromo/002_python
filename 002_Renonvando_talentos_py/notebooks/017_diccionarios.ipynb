

Lo que esta antes de los : es la key y lo que está depués son los elementos
Dentro de una lista o un diccionario pòdemos poner una cadena poniendo : 

Los diccionarios sirven para acceder a un elemento poniendo una key

```python
diccionario = {"nombre":"Erick",
              "apellido":"Romo",
              "edad":23}
type(diccionario)
```

Podemos acceder o reemplazar elementos poniendo dict[key] = algo

```python
diccionario["apellido"]
```

```python
diccionario["apellido"] = "Eligio"
diccionario["apellido"]
```

Podemos agreagar más llaves y elementos con dict[key] = algo

```python
diccionario["primer_apellido"]="romo"
diccionario
```

```python
diccionario.pop("primer_apellido")
diccionario
```

```python
diccionario["primer_apellido"]="romo"
diccionario
```

Tambi+en se puede borrar con del diccionario["key"]

```python
del diccionario["primer_apellido"]
```
