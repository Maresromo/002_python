```python
import pandas as pd
```

```python
f = "../data/esolmet.txt"            #Editando con notepad++ se sustituyo ", " por " ", usando este ultimo como separador
esolmet = pd.read_csv(f,sep=" ",     #Después, se sustituyó "," por "" para evitar que se reconozcan como objetos
                      index_col=0,
                      parse_dates=True)     
esolmet.sort_index(inplace = True)
esolmet
```

```python
esolmet.to_csv("../data/esolmetFORMATO.csv", index=True)  # Usa index=True si quieres incluir los índices en el CSV
```
