```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse
```

```python
f = '../data/Temixco_2018_10Min.csv'
tmx = pd.read_csv(f,index_col=0,parse_dates=True)
tmx
```

```python
fig, ax = plt.subplots(2, figsize=(12,4), sharex=True)

f1 = parse("2018-01-13")
f2 = f1 + pd.Timedelta('3D')

ax[0].plot(tmx.Ig, label='Ig')
ax[1].plot(tmx.To, label='To')

ax[0].legend()
ax[1].legend()

ax[0].set_xlim(f1,f2)
```

```python
columnas=tmx.columns
Is = [columna for columna in columnas if "I" in columna]
Is   #Is es una lista que contiene los nombres de las series que queremos visualizar, es util para automatizar
```

```python
fig, ax = plt.subplots(len(Is), figsize=(10,4), sharex=True)   #con len(Is) nos cuenta cuantas series vamos a graficar

for i,I in enumerate(Is):        #con enumerate nos da i= indice, I=nombre del objeto
    ax[i].plot(tmx[I],label=I)

for eje in ax:                   #Aca graficamos todas las series que tenemos en ax
    eje.legend()
```
